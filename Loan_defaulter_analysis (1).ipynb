```python
plt.figure(figsize=(25,60))
for i, var in enumerate(object_variable):
    print(f"Processing: {var}")

    try:
        data_pct = app_ext_rmvd[[var,'TARGET']].groupby([var], as_index=False).mean().sort_values(by='TARGET',ascending=False)
        data_pct['PCT'] = data_pct['TARGET']*100

        plt.subplot(10,2,i+i+1)
        plt.subplots_adjust(wspace=0.1,hspace=1)
        sns.countplot(data=app_ext_rmvd,x=var,hue='TARGET')
        plt.xticks(rotation=90)

        plt.subplot(10,2,i+i+2)
        sns.barplot(data=data_pct, x=var, y='PCT', hue=var, palette='coolwarm', legend=False)
        plt.xticks(rotation=90)

    except Exception as e:
        print(f"Skipping {var} due to error: {e}")
```
